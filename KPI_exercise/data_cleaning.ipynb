```python
import pandas as pd
from datetime import date
from dateutil.relativedelta import relativedelta
```

```python
df1 = pd.read_csv('/Users/yairmartinez/Documents/GitHub/Data_analysis_exercises/DB/Datos/application_record.csv')
df2 = pd.read_csv('/Users/yairmartinez/Documents/GitHub/Data_analysis_exercises/DB/Datos/credit_record.csv')
```

```python
### Data cleaning

## df1 a df3

# Definición df3 (df1 limpio)
df3 = df1.assign(DAYS_BIRTH= lambda x: round(   ( (df1['DAYS_BIRTH'] *-1))/365   , 2  )   ,
DAYS_EMPLOYED=lambda x: round(  ( (df1['DAYS_EMPLOYED'] *-1)) / 365      , 2  )          )

# Renombrar columnas df3
df3.rename(columns = {'DAYS_BIRTH':'YEARS_BIRTH'}, inplace = True)
df3.rename(columns = {'DAYS_EMPLOYED':'YEARS_EMPLOYED'}, inplace = True)


## df2 a df4

# Renombrar columna del df4
df4 = df2
hoy=date.today()
lista_mes_balance = []

for x in df4['MONTHS_BALANCE']:
    mes_balance = hoy + relativedelta(months=x)
    lista_mes_balance.append(mes_balance)

df4['MONTHS_BALANCE'] = lista_mes_balance
```

```python
print(df3.shape,df4.shape)
```

```python
dfAll = pd.merge(df3, df4, how='inner', on=['ID'])
```
