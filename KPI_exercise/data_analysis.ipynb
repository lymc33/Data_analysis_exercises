```python
# Analisis de clientes buenos y malos

# MES ACTUAL

mes_actual= dfAll[ ( dfAll['MONTHS_BALANCE'] == 0 )   ]
mes_actual.groupby(['STATUS']).describe( )
```

```python
# Analisis de clientes buenos y malos
# MES -1  =  mes_m1

mes_m1= dfAll[ ( dfAll['MONTHS_BALANCE'] == -1 )   ]
mes_m1.groupby(['STATUS']).describe( )
```

```python
# Analisis de clientes buenos y malos
# mes_actual, conteo de claves de status

mes_actual[ ( mes_actual['MONTHS_BALANCE'] == 0 )   ].groupby(['STATUS']).count()
```
